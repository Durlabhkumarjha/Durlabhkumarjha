```python
# Importing Libararies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Importing the Dataset
lead_data=pd.read_csv("train.csv")

lead_data.head()
```

```python
# Lets check the dimensions of Dataframe
lead_data.shape
```

```python
lead_data.info()
```

```python
# let's look at the statistical aspects of the dataframe
lead_data.describe()
```

```python
lead_data.isna().sum().sort_values(ascending=False).head(20)
```

```python
lead_data['PoolQC']=lead_data['PoolQC'].fillna('nopool')
lead_data['MiscFeature']=lead_data['MiscFeature'].fillna('none')
lead_data['Alley']=lead_data['Alley'].fillna('noAlleyaccess')
lead_data['Fence']=lead_data['Fence'].fillna('noFence')
lead_data['FireplaceQu']=lead_data['FireplaceQu'].fillna('noFireplace')
```

```python
round((lead_data.isna().sum().sort_values(ascending=False)/lead_data.shape[0])*100,2).head(15)
```

```python
lead_data=lead_data.dropna(how='all')
```

```python
round((lead_data.isna().sum().sort_values(ascending=False)/lead_data.shape[0])*100,2).head(15)
```

```python
lead_data['LotFrontage'].value_counts()
```

```python
numerical_variable=lead_data.describe().columns
categorical_variable=lead_data.select_dtypes('object').columns
len(categorical_variable)
len(numerical_variable)
```

```python

#checking the uniqueness of the categorical values
for i in lead_data.columns:
    print(i,lead_data[i].nunique(dropna=True))
```

```python
# Finding the unique values
for i in categorical_variable:
    plt.pie(x=lead_data[i].value_counts(),labels=lead_data[i].value_counts().index,autopct='%.0f%%')
    plt.title(i)
    plt.show()
```
